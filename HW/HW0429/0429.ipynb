```python
import mglearn
import matplotlib.pyplot as plt
```

```python
# generate dataset
X, y = mglearn.datasets.make_forge()
# plot dataset
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0", "Class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print("X.shape:", X.shape)
```

```python
X, y = mglearn.datasets.make_wave(n_samples=40)
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel("Feature")
plt.ylabel("Target")
```

```python
from sklearn.datasets import load_breast_cancer
import numpy as np
cancer = load_breast_cancer()
print("cancer.keys():\n", cancer.keys())
print("Shape of cancer data:", cancer.data.shape)
print("Sample counts per class:\n",
      {n: v for n, v in zip(cancer.target_names, np.bincount(cancer.target))})
print("Feature names:\n", cancer.feature_names)
```

```python
from sklearn.datasets import load_wine
import numpy as np
wine = load_wine()
print("wine.keys():\n", wine.keys())
print("Shape of wine data:", wine.data.shape)
print("Sample counts per class:\n",
      {n: v for n, v in zip(wine.target_names, np.bincount(wine.target))})
print("Feature names:\n", wine.feature_names)
```

```python
features_mean = []
```

```python
for col in data.columns:
    if 'mean' in col:
        features_mean.append(col)
```

```python
np.shape(features_mean)
```

```python
plt.figure(figsize=(28,28))
sns.heatmap(data[features_mean].corr(), annot=True, square=True, cmap='coolwarm')
plt.show()
```

```python
color_dic = {'M':'red', 'B':'blue'}
colors = data['label'].map(lambda x: color_dic.get(x))

sm = pd.plotting.scatter_matrix(data[features_mean], c=colors, alpha=0.4, figsize=((15,15)));

plt.show()
```
